# שקף 8

```python
import random
```

```python
def roll_dices():
    a = random.randint(1, 6)
    b = random.randint(1, 6)
    c = random.randint(1, 6)
    d = random.randint(1, 6)
    e = random.randint(1, 6)
    return a, b, c, d, e
```

# שקף 9

```python
player = roll_dices()
computer = roll_dices()
print(f"The computer's roll: {computer}")
print(f"The player's roll: {player}")
winner = 'player' if sum(player) > sum(computer) else 'computer'
print(f"The winner is {winner}")
```

```python
# getting a string of numbers and returning hist:
def hist_fun(string):
    hist = [0] * 10
    for s in string:
        hist[int(s)] += 1
    return hist

# check_string:
def check_string(string):
    hist = hist_fun(string)
    for i, val in enumerate(hist):
        if i == val or val == 0:
            continue
        return False
    return True
```
